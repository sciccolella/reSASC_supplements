```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
def load_data(exp_no):
    df_ad = pd.read_csv(f'plots/exp{exp_no}.anc_desc.csv')
    df_dl = pd.read_csv(f'plots/exp{exp_no}.diff_lin.csv')
    df_mp3 = pd.read_csv(f'plots/exp{exp_no}.mp3.csv')
    df_lj = pd.read_csv(f'plots/exp{exp_no}.loss_jaccard.csv')
    df_rj = pd.read_csv(f'plots/exp{exp_no}.rec_jaccard.csv')
    df_ll = pd.read_csv(f'plots/exp{exp_no}.loglikelihood.csv')

    #re-order to keep color consistency
    df_ll = df_ll[["reSASC", "Dollo-k", "PP", "Ground", "Measure"]]

    return df_ad, df_dl, df_mp3, df_lj, df_rj, df_ll
    # return df_ad, df_dl, df_mp3, df_ll
```

```python
def boxplot(dfs, out=None):
    f, ax = plt.subplots(1, len(dfs), figsize=(4*len(dfs),4))
    for ix, df in enumerate(dfs):
        # _ix = np.unravel_index(ix, (2, len(dfs)//2))
        _ix = ix
        ax[_ix] = sns.boxplot(data=df, ax=ax[_ix])
        ax[_ix] = sns.stripplot(data=df, ax=ax[_ix],alpha=.6, edgecolor="gray", linewidth=.5)
        ax[_ix].set(title=df["Measure"][0])
    plt.tight_layout()
    if out:
        f.savefig(out)
    return ax
```

```python
for exp in range(1, 5): # 1, 6
    boxplot(load_data(exp), out=f'plots/exp{exp}.pdf')
```
